## Preprocessing (Handling Missing values)

```python
df["Age"].unique()[:300]
```

```python
cols=["Hypertension","Diabetes_mellitus","Coronary_artery_disease","Peda_edema","Appetite","Aanemia"]
df.dropna(subset=cols, inplace=True)


df.isnull().sum()
```

```python
Age_mean =df["Age"].mean()
Age_mean
```

```python
df["Age"]=df["Age"].fillna(Age_mean)
```

```python
col=["Blood_pressure","Pus_cell_clumps","Bacteria","Blood_urea","Serum_creatinine"]
df[col]= df[col].ffill()
```

```python
df.isnull().sum()
```

```python
col1=['Sodium','Potassium','Haemoglobin','Sugar','Packed_cell_volume','albumin','White_blood_cell_count','Red_blood_cell_count','Specific_gravity','Blood_glucose_random',]
col1_mean=df[col1].mean()
col1_mean
```

```python
df[col1]=df[col1].fillna(col1_mean)
```

```python
df.isnull().sum()
```

```python
col2=['Pus_cell','Red_blood_cells']
col2_mode= df[col2].mode().iloc[0]
col2_mode
```

```python
df[col2]=df[col2].fillna(col2_mode)
```

```python
df.isnull().sum()
```

## Building Logistic Regression Model

### Target variables

```python
ind_cols=[col for col in df.columns if col!="Class"] ##select every column as a fearture col except Class
dep_col = "Class"

X=df[ind_cols]
y=df[dep_col]
```

### Spliting Datasets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.3,random_state=0)
```

### Training Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

Lr = LogisticRegression()

Lr.fit(X_train,y_train) ##fitting the model

y_pred =Lr.predict(X_test) ## Doing prediction
## Evaluation
Accuracy = accuracy_score(y_test,y_pred)
print(f"Training Accuracy of Logistic Regression is {accuracy_score(y_train, Lr.predict(X_train))}")


print((f"Model Testing Accuracy is:{Accuracy}"))

print(f"Confusion Matrix :- \n{confusion_matrix(y_test, Lr.predict(X_test))}\n")

```

```python

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Predicted 0", "Predicted 1"],
    yticklabels=["Actual 0", "Actual 1"]
)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Kidney Disease - Logistic Regression")
plt.show()
```
